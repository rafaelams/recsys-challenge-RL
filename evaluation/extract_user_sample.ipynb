```python
import pandas as pd
import numpy as np
```

```python
df_review = pd.read_csv("../data/dataset/yelp_dataset/yelp_academic_dataset_review.csv")
df_review = df_review[df_review["stars"] >= 4]
df_review.head()
```

```python
df_review.shape 
```

```python
df_bussines = pd.read_csv("../data/dataset/yelp_dataset/yelp_academic_dataset_business.csv")[["business_id", "city"]]
bussines_city = df_bussines.groupby("city").agg({ "business_id": list}).to_dict()["business_id"]
```

```python
df_review_total = df_review\
                    .groupby("user_id").agg(
                      total_reviews = ("business_id", "count")
                    ).reset_index()
df_review_total.head()
```

```python
df_review_with_5 = df_review[df_review['stars'] >= 5]\
                    .groupby("user_id").agg(
                      business_with_5 = ("business_id", list), 
                      total_with_5 = ("business_id", "count")
                    ).reset_index()
df_review_with_5.head()
```

```python
df_users = df_review_total\
            .merge(df_review_with_5, on="user_id", how="inner")

# Filter only users with 5 or more reviews
df_users = df_users[df_users['total_reviews'] == 10].sample(1000, random_state=42)
df_users.shape
```

```python
df_users
```

```python
df_eval_review = df_review\
                .groupby("user_id").agg(
                  business_ids = ("business_id", list), 
                  business_total = ("business_id", "count"),
                  business_id = ("business_id", lambda x: x.sample(1).values[0])
                )\
                .merge(df_users, on="user_id", how="inner")\
                .merge(df_bussines, on="business_id", how="inner")\
                .reset_index()

df_eval_review.head()
```

```python
df_eval_review["business_random"] = df_eval_review["city"].apply(lambda x: list(np.unique(np.random.choice(bussines_city[x], 10, replace=True))))
df_eval_review.head()
```

```python
# join two columns lists in dataframe
import random

df_eval_review["gt_reclist"] = df_eval_review.apply(lambda x: x.business_ids, axis=1)
df_eval_review["user_perfil"] = df_eval_review.apply(lambda x: x.business_ids[0], axis=1)
df_eval_review["reclist"] = df_eval_review.apply(lambda x: x.business_ids[1:-1] + x.business_random, axis=1)
df_eval_review["reclist"] = df_eval_review.apply(lambda x: random.sample(x.reclist, len(x.reclist)), axis=1)
df_eval_review["total_reclist"] = df_eval_review["reclist"].apply(len)
df_eval_review.head()
```

```python
df_eval_review[["user_id", "user_perfil", "gt_reclist", "reclist"]]\
  .to_csv("../data/evaluation/eval_users.csv", index=False)
```
